La fonction $f(x)=x^3$ verifie $f'(0)=0$, mais $x=0$ n'est pas un extremum local. En effet, $f(x)<0$ pour $x<0$ et $f(x)>0$ pour $x>0$.

## Question 1 (M + C)

On prend $X=0$ presque surement et $Y\sim \mathrm{Exp}(1)$.

Pour la perte quadratique, le predicteur de Bayes est la moyenne:

$$f^*_{squared}(0)=\mathbb E[Y]=1.$$

Pour la perte absolue, le predicteur de Bayes est une mediane conditionnelle. Pour une loi exponentielle de parametre 1, la mediane vaut:

$$f^*_{absolute}(0)=\log(2).$$

Ces deux predicteurs sont differents.

```python
import numpy as np

rng = np.random.default_rng(123)
n_test = 500_000
y_test = rng.exponential(scale=1.0, size=n_test)

f_squared = np.ones_like(y_test)
f_absolute = np.full_like(y_test, np.log(2))

risk_abs_squared = np.mean(np.abs(y_test - f_squared))
risk_abs_absolute = np.mean(np.abs(y_test - f_absolute))

print(f"Predicteur perte quadratique: {1.0:.4f}")
print(f"Predicteur perte absolue:     {np.log(2):.4f}")
print(f"Risque absolu du predicteur quadratique: {risk_abs_squared:.4f}")
print(f"Risque absolu du predicteur absolu:      {risk_abs_absolute:.4f}")
print(f"Le predicteur absolu est meilleur: {risk_abs_absolute < risk_abs_squared}")
```

## Question 2 (M)

On fixe $x$ et on note $p = p_{Y\mid X=x}$ la densite conditionnelle (continue) et $F_x(z)=\int_{-\infty}^{z}p(y)dy$ la fonction de repartition conditionnelle de $Y\mid X=x$. On veut minimiser sur $\mathbb R$:

$$g(z)=\int_{\mathbb R}|y-z|\,p(y)\,dy.$$

**Existence et finitude.** Comme $Y\mid X=x$ admet un moment d'ordre 1, $g(z) \le \mathbb E[|Y|\mid X=x] + |z| < +\infty$ pour tout $z$: $g$ est bien definie sur $\mathbb R$.

**Convexite.** Pour chaque $y$ fixe, $z\mapsto |y-z|$ est convexe. Une integrale (a poids positif $p(y)dy$) de fonctions convexes est convexe, donc $g$ est convexe sur $\mathbb R$.

Ce point est crucial a cause du piege illustre en Question 0: pour une fonction quelconque, annuler la derivee ne suffit pas a garantir un extremum. Pour une fonction **convexe**, en revanche, tout point critique est un minimum global.

**Calcul de la derivee.** On separe l'integrale en $y \le z$ et $y > z$:

$$g(z)=\int_{-\infty}^{z}(z-y)p(y)dy + \int_z^{+\infty}(y-z)p(y)dy.$$

On derive chaque terme par rapport a $z$ (derivation sous l'integrale, licite car $|y-z|$ est 1-lipschitzienne en $z$ et $p$ integrable; les termes de bord en $y=z$ s'annulent car l'integrande y vaut $0$):

$$g'(z)=\int_{-\infty}^{z}p(y)dy - \int_z^{+\infty}p(y)dy = F_x(z)-\big(1-F_x(z)\big) = 2F_x(z)-1.$$

**Conclusion.** $g$ est convexe et derivable, donc $z^*$ est un minimiseur global si et seulement si $g'(z^*)=0$, c'est-a-dire:

$$F_x(z^*)=\frac12.$$

Le predicteur de Bayes pour la perte absolue est donc une **mediane conditionnelle** de $Y\mid X=x$:

$$f^*_{absolute}(x) \in \{z\in\mathbb R:\ F_{Y\mid X=x}(z)=1/2\}.$$

Comme $F_x$ est continue (densite continue), croissante, de limites $0$ et $1$, un tel $z^*$ existe toujours; il est unique des que $p>0$ au voisinage de la mediane. Ceci est coherent avec la Question 1: pour $\mathrm{Exp}(1)$, $F(z)=1-e^{-z}=1/2$ donne $z=\log 2$, distinct de la moyenne $1$.

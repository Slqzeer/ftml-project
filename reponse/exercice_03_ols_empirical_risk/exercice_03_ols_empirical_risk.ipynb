# Exercice 3 - Expected value of empirical risk for OLS

## Cadre

Modele lineaire a design fixe:

$$y=X\theta^*+\varepsilon,\qquad \varepsilon\sim\mathcal N(0,\sigma^2 I_n),$$

avec $X\in\mathbb R^{n\times d}$ deterministe et injective (donc $X^TX$ inversible et $d\le n$). L'estimateur OLS est $\hat\theta=(X^TX)^{-1}X^Ty$.

## Question 1 (M) - comparaison a un risque de Bayes

Pour un point fixe $x_i$ du design, le predicteur de Bayes pour la perte quadratique est $f^*(x_i)=\mathbb E[y_i]=\theta^{*T}x_i$, et le risque de Bayes vaut

$$R^* = \mathbb E[(y_i-\theta^{*T}x_i)^2]=\mathbb E[\varepsilon_i^2]=\sigma^2.$$

La Proposition 1 donne $\mathbb E[R_X(\hat\theta)]=\frac{n-d}{n}\sigma^2 < \sigma^2 = R^*$ des que $d\ge 1$.

Interpretation: le risque empirique de l'OLS est mesure sur les donnees qui ont servi a ajuster $\hat\theta$; il est donc **optimiste**: en moyenne il est meme inferieur au risque de Bayes, ce qu'aucun estimateur ne peut atteindre en generalisation. Chaque degre de liberte de $\theta$ absorbe une part $\sigma^2/n$ du bruit:

- si $n\gg d$, $(n-d)/n\approx 1$ et le risque empirique est un bon estimateur de $\sigma^2$;
- si $d\to n$, $(n-d)/n\to 0$: le modele interpole le bruit et le risque empirique devient artificiellement nul, sans que la generalisation s'ameliore.

## Question 2 (M)

Le vecteur des residus s'ecrit, en substituant $y=X\theta^*+\varepsilon$ dans $\hat\theta$:

$$\hat\theta=(X^TX)^{-1}X^T(X\theta^*+\varepsilon)=\theta^*+(X^TX)^{-1}X^T\varepsilon.$$

Donc

$$
y-X\hat\theta = X\theta^*+\varepsilon-X\theta^*-X(X^TX)^{-1}X^T\varepsilon
=\big(I_n-X(X^TX)^{-1}X^T\big)\varepsilon.
$$

Ainsi, avec $A=I_n-X(X^TX)^{-1}X^T$:

$$\mathbb E\big[R_n(\hat\theta)\big]=\mathbb E\Big[\frac1n\|y-X\hat\theta\|^2\Big]=\mathbb E_\varepsilon\Big[\frac1n\|A\varepsilon\|^2\Big],$$

ou l'esperance porte uniquement sur $\varepsilon$, seule source d'alea en design fixe. C'est l'egalite (9) demandee.

## Question 3 (M)

Soit $A\in\mathbb R^{n\times n}$. Par definition du produit matriciel, pour tout $j$:

$$(A^TA)_{jj}=\sum_{i=1}^n (A^T)_{ji}A_{ij}=\sum_{i=1}^n A_{ij}^2.$$

En sommant sur $j$:

$$\mathrm{tr}(A^TA)=\sum_{j=1}^n (A^TA)_{jj}=\sum_{j=1}^n\sum_{i=1}^n A_{ij}^2=\sum_{(i,j)\in[1,n]^2}A_{ij}^2.$$

## Question 4 (M)

On developpe $\|A\varepsilon\|^2$:

$$\|A\varepsilon\|^2=(A\varepsilon)^T(A\varepsilon)=\varepsilon^TA^TA\varepsilon=\sum_{j,k}(A^TA)_{jk}\,\varepsilon_j\varepsilon_k.$$

Par linearite de l'esperance et puisque les $\varepsilon_j$ sont centres, independants, de variance $\sigma^2$:

$$\mathbb E[\varepsilon_j\varepsilon_k]=\sigma^2\,\delta_{jk}
\quad\Longrightarrow\quad
\mathbb E\big[\|A\varepsilon\|^2\big]=\sum_{j}(A^TA)_{jj}\,\sigma^2=\sigma^2\,\mathrm{tr}(A^TA).$$

D'ou

$$\mathbb E_\varepsilon\Big[\frac1n\|A\varepsilon\|^2\Big]=\frac{\sigma^2}{n}\mathrm{tr}(A^TA).$$

## Question 5 (M)

Notons $H=X(X^TX)^{-1}X^T$ et $A=I_n-H$.

- **$H$ est symetrique**: $H^T=\big(X(X^TX)^{-1}X^T\big)^T=X\big((X^TX)^{-1}\big)^TX^T=X\big((X^TX)^T\big)^{-1}X^T=X(X^TX)^{-1}X^T=H$, car $X^TX$ est symetrique.
- **$H$ est idempotente**: $H^2=X(X^TX)^{-1}\underbrace{X^TX(X^TX)^{-1}}_{=I_d}X^T=X(X^TX)^{-1}X^T=H$.

Alors $A^T=I_n-H^T=I_n-H=A$ et

$$A^TA=A^2=(I_n-H)^2=I_n-2H+H^2=I_n-2H+H=I_n-H=A.$$

## Question 6 (M) - conclusion

$H$ est la projection orthogonale sur $\mathrm{Im}(X)$, sous-espace de dimension $d$ ($X$ injective), donc $\mathrm{tr}(H)=d$ (la trace d'un projecteur est son rang, ou bien: $\mathrm{tr}(H)=\mathrm{tr}\big(X(X^TX)^{-1}X^T\big)=\mathrm{tr}\big((X^TX)^{-1}X^TX\big)=\mathrm{tr}(I_d)=d$ par propriete cyclique de la trace).

En combinant les questions 2, 4 et 5:

$$\mathbb E\big[R_n(\hat\theta)\big]
=\frac{\sigma^2}{n}\mathrm{tr}(A^TA)
=\frac{\sigma^2}{n}\mathrm{tr}(A)
=\frac{\sigma^2}{n}\big(\mathrm{tr}(I_n)-\mathrm{tr}(H)\big)
=\frac{n-d}{n}\sigma^2.$$

Ce qui prouve la Proposition 1.

## Question 7 (M)

Comme $R_n(\hat\theta)=\|y-X\hat\theta\|_2^2/n$:

$$\mathbb E[\|y-X\hat\theta\|_2^2]=(n-d)\sigma^2.$$

Donc:

$$\mathbb E\left[\frac{\|y-X\hat\theta\|_2^2}{n-d}\right]=\sigma^2.$$

```python
import numpy as np

rng = np.random.default_rng(7)
n, d = 120, 20
sigma = 1.7
n_rep = 3000

X = rng.normal(size=(n, d))
theta_star = rng.normal(size=d)

sigma2_hats = []
for _ in range(n_rep):
    eps = rng.normal(0, sigma, size=n)
    y = X @ theta_star + eps
    theta_hat = np.linalg.lstsq(X, y, rcond=None)[0]
    residuals = y - X @ theta_hat
    sigma2_hats.append(np.sum(residuals**2) / (n - d))

sigma2_hats = np.array(sigma2_hats)
print(f"sigma^2 theorique: {sigma**2:.4f}")
print(f"moyenne estimee:   {sigma2_hats.mean():.4f}")
print(f"ecart-type estim.: {sigma2_hats.std():.4f}")
```

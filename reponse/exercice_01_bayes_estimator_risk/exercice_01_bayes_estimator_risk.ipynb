# Exercice 1 - Bayes estimator and Bayes risk

Objectif: proposer un cadre supervise avec une quantite physique, calculer le predicteur de Bayes et verifier par simulation qu'il bat un estimateur volontairement moins bon.

## Question 1 (M)

On modelise des ventes journalieres de glace en fonction de la temperature exterieure.

- Espace d'entree: $\mathcal X = [0, 30]$, temperature moyenne du jour en degres Celsius.
- Espace de sortie: $\mathcal Y = \mathbb R_+$, volume vendu dans la journee, en litres.
- Loi: $X \sim \mathcal U([0,30])$ et

$$Y = 10 + 2X + \varepsilon,$$

avec $\varepsilon \sim \mathcal N(0, 2^2)$ independant de $X$.

On utilise la perte quadratique:

$$\ell(y,z) = (y-z)^2.$$

Pour la perte quadratique, le predicteur de Bayes est l'esperance conditionnelle:

$$f^*(x) = \mathbb E[Y \mid X=x] = 10 + 2x.$$

Le risque de Bayes vaut:

$$R(f^*) = \mathbb E[(Y-f^*(X))^2] = \mathbb E[\varepsilon^2] = 4.$$

L'unite du risque est le litre carre, car la perte est une erreur quadratique sur un volume en litres.

## Question 2 (C)

On compare $f^*(x)=10+2x$ a un estimateur non optimal choisi manuellement:

$$\hat f(x)=15+1.6x.$$

Son risque theorique se calcule aussi: en notant $b(x)=f^*(x)-\hat f(x)=0.4x-5$,

$$R(\hat f)=\mathbb E[(Y-\hat f(X))^2]=\mathbb E[\varepsilon^2]+\mathbb E[b(X)^2]=4+\mathbb E[(0.4X-5)^2].$$

Avec $X\sim\mathcal U([0,30])$, $0.4X-5\sim\mathcal U([-5,7])$, donc $\mathbb E[(0.4X-5)^2]=\mathrm{Var}+\mathrm{moyenne}^2=\frac{12^2}{12}+1^2=13$ et $R(\hat f)=17$.

La simulation estime le risque reel par un risque empirique sur un grand test set independant.

```python
import numpy as np

rng = np.random.default_rng(42)
n_test = 200_000

X_test = rng.uniform(0, 30, size=n_test)
eps = rng.normal(0, 2, size=n_test)
y_test = 10 + 2 * X_test + eps


def f_bayes(x):
    return 10 + 2 * x


def f_hat(x):
    return 15 + 1.6 * x


risk_bayes = np.mean((y_test - f_bayes(X_test)) ** 2)
risk_hat = np.mean((y_test - f_hat(X_test)) ** 2)

print("Bayes risk theorique:      4.000")
print("Risque theorique de f_hat: 17.000")
print(f"Risque test f*:            {risk_bayes:.3f}")
print(f"Risque test f_hat:         {risk_hat:.3f}")
print(f"f* est meilleur:           {risk_bayes < risk_hat}")
```

Conclusion: le risque test de $f^*$ est proche de la valeur theorique 4 et celui de $\hat f$ est proche de sa valeur theorique 17, lorsque le test set est grand. L'estimateur $\hat f$ est moins bon car il a une pente et une ordonnee a l'origine biaisees, et le risque de Bayes est bien le plus petit risque atteignable.
